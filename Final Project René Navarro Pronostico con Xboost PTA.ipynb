```python
def create_features(pta1):
        pta1['quarter'] = pta1.index.quarter
        pta1['month'] = pta1.index.month
        pta1['year'] = pta1.index.year
        return pta1
pta1 = create_features(pta1)
```

```python
fig, ax = plt.subplots(figsize=(10, 8))
sns.boxplot(data=pta1, x='month', y='PTA c/lb')
ax.set_title('Price by Month')
plt.show()
```

##### Modelo

##### Modelo para PTA

```python
pta1_train = create_features(pta1_train)
pta1_test = create_features(pta1_test)

FEATURES = ['quarter', 'month', 'year']
TARGET = 'PTA c/lb'

X_train = pta1_train[FEATURES]
y_train = pta1_train[TARGET]

X_test = pta1_test[FEATURES]
y_test = pta1_test[TARGET]
```

```python
reg = xgb.XGBRegressor(base_score=0.5, booster='gbtree',    
                       n_estimators=1000,
                       early_stopping_rounds=50,
                       objective='reg:linear',
                       max_depth=3,
                       learning_rate=0.01)
reg.fit(X_train, y_train,
        eval_set=[(X_train, y_train), (X_test, y_test)],
        verbose=100)
```

```python
fi = pd.DataFrame(data=reg.feature_importances_,
             index=reg.feature_names_in_,
             columns=['importance'])
fi.sort_values('importance').plot(kind='barh', title='Feature Importance')
plt.show()
```

```python
pta1_test['prediction'] = reg.predict(X_test)
pta1 = pta1.merge(pta1_test[['prediction']], how='left', left_index=True, right_index=True)
ax = pta1[['PTA c/lb']].plot(figsize=(15, 5))
pta1['prediction'].plot(ax=ax, style='.')
plt.legend(['Truth Data', 'Predictions'])
ax.set_title('Raw Dat and Prediction')
plt.show()
```

```python
score = np.sqrt(mean_squared_error(pta1_test['PTA c/lb'], pta1_test['prediction']))
print(f'RMSE Score on Test set: {score:0.2f}')
```

```python
pta1_test['error'] = np.abs(pta1_test[TARGET] - pta1_test['prediction'])
pta1_test['date'] = pta1_test.index.date
pta1_test.groupby(['date'])['error'].mean().sort_values(ascending=False).head(10)
```
